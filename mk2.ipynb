МК 2. Тиждень 26.04.2021 - захист

На внутрішньому ринку популярні піцерії пропонують до 15 **видів** піц, кожна із яких формується із
різних інгредієнтів, у залежності від рецепту - див. стовпчик МК-2 результатів захисту індивідуальних завдань і МК. 
Ваша піцерія організує автоматизацію обслуговування клієнтів, у залежності від власного
варіанту, який вказано у стовпчику МК-2 результатів захисту індивідуальних завдань і МК і відповідає ПІБ студента. Для кожного
студента із загального списку популярних піц, наданого за посиланням, для виконання МК-2 надано
лише 5 **видів** піц із відповідними номерами (назвами піц).
На основі принципів ООП, необхідно розробити і реалізувати застосунок, який забезпечує автоматизоване
обслуговування клієнтів, бажаючих замовити, в режимі самообслуговування, піцу із вказаного списку із 5 піц (згідно власного варіанту).
У процесі обслуговування клієнта необхідно використовувати наступні об'єкти (класи):
- МЕНЮ. Відповідальність об'єкту: взаємодія з клієнтом, вивід меню піцерії з
виводом назви піці і її номеру із загального списку, підтвердження замовлення піці із
вказаного меню, оплата з власного банківського рахунку. Для оплати замовлених піц (ЗАМОВЛЕННЯ) використовувати банківський акаунт, розроблений
в МК-1. Причому, наявними є три **типи** піці: стандарт, маленька і велика. Вага стандартної піці і її інгредієнтів задається виконавцем.
Причому, собівартість маленької піці 60% від стандартної піці з націнкою 5%, а собівартість великої піці 140% від стандартної піці з дисконтом 5% для кожного **виду** піці.
Собівартість кожного інгредієнту великої і маленької піці, розраховується на основі собівартості кожного інгредієнту піці стандарт, які задаються виконавцем.   
- ЗАМОВЛЕННЯ. Відповідальність об'єкту: містить список замовлених піц (не менше 3-х) у відповідності до собівартості
її інгредієнтів. 
- ПІЦА. Відповідальність об'єкту: базовий клас піц, відповідає заявленим узагальненим характеристикам
піц власного варіанту. 
- ПІЦА_назва+№. Відповідальність об'єкту: успадкований клас абстрактного базового класу ПІЦА, в якому здійснюється уточнення 
характеристик (атрибутів, інгредієнтів і методів) кожного **виду** піці. Ім'я кожного успадкованого класу ПІЦА_назва+№ відповідає назві і номеру піці, згідно власного варіанту.
Таких успадкованих класів піц - 5, згідно номерів **видів** піц власного варіанту. 

## Завдання:


1. Автоматизований термінал обслуговування клієнта повинен відображати `МЕНЮ` піц із 5 піц власного варіанту.  
2. МЕНЮ дозволяє в автоматизованому варіанті створити `ЗАМОВЛЕННЯ`, з урахуванням **типу** і **виду** піці. 
3. Клієнт формує `ЗАМОВЛЕННЯ` шляхом введення номера піці із `МЕНЮ`.
4. `ЗАМОВЛЕННЯ` формується у вигляді списку із номерів піц, представлених в `МЕНЮ`.
5. У `ЗАМОВЛЕННІ` повинно бути не менше 3-х **видів** піц обраного типу.
6. Клієнт підтверджує `ЗАМОВЛЕННЯ` обраних піц або може відмінити `ЗАМОВЛЕННЯ`.
7. Клієнт отримує рахунок за `ЗАМОВЛЕННЯ` і може перевірити `СТАТУС` банківського рахунку до оплати `ЗАМОВЛЕННЯ` із автоматизованим визначенням можливості оплати цього `ЗАМОВЛЕННЯ`.
8. Клієнт сплачує `ЗАМОВЛЕННЯ` з власного банківського рахунку.
9.  Клієнт отримує чек про оплату `ЗАМОВЛЕННЯ`: **типи** і **види** замовлених піц, **ціна** кожної і загальна **сума**.
10. Клієнт отримує статус власного банківського рахунку після оплати `ЗАМОВЛЕННЯ`.
