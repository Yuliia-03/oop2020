```python
from menu import Menu
from coffee_maker import CoffeeMaker
from money_machine import MoneyMachine

menu = Menu()
money_machine = MoneyMachine()
coffee_maker = CoffeeMaker()



while True:
    
    order = input("Яку каву бажаєте? {}:".format(menu.get_items()))
    
    
    if order == 'дякую':
        print('Апарат завершив роботу!')
        break
    
    elif order == 'чек':
        money_machine.report()
        coffee_maker.report()
    
    elif menu.find_drink(order) != None:
        price = menu.find_drink(order).cost
        
        if coffee_maker.is_resource_sufficient(menu.find_drink(order)):
            print("Ціна обраної кави:", price)
            
            if money_machine.make_payment(price):
                coffee_maker.make_coffee(menu.find_drink(order))
```
