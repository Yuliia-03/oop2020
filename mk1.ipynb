```python
class MyAccount:
    
    
    def set_account(self, first_name, second_name, status = ""):
        
        self.first_name = first_name
        self.second_name = second_name
        self.amound_of_money = 0
        self.deposit_money = 0
        
        self.status = status
        
    def my_open_account(self):
        
        self.status = str(self.first_name + " " + self.second_name)
        self.status += str("\nБаланс рахунку: " + str(self.amound_of_money))
        self.status += str("\nБаланс депозитних коштів: " + str(self.deposit_money) + "\n")
        return self.status

    def put_account(self, money):
        
        self.amound_of_money += money
        return self.amound_of_money
       
    def get_accaunt(self, money):
        
        if self.amound_of_money >= money:
            self.amound_of_money -= money
        else:
            print("Недостатньо коштів на балансі")
        
        return self.amound_of_money
    
    
    def send_deposit(self, amound_of_money, deposit_rate, deposit_time):
        
        self.deposit_money = (amound_of_money * (deposit_rate/100) / 12) * deposit_time
        self.tax = self.deposit_money * 0.18
        self.deposit_money -= self.tax
        self.deposit_money += amound_of_money
        return self.deposit_money
    
    

account1 = MyAccount()
account2 = MyAccount()

account1.set_account("Сидоренко", "Олеся")
account2.set_account("Федоренко", "Олександр")


account1.put_account(1000)
account2.put_account(2000)

account1.get_accaunt(100)
account2.get_accaunt(200)

account1.send_deposit(500, 5, 30)
account2.send_deposit(100, 10, 6)

print(account1.my_open_account())
print(account2.my_open_account())
```

Припустимо, що власник придбав зі знижкою вказані автомобілі завдання 9 на первинному ринку. Після року експлуатації всіх чотирьох автомобілів, власник вирішив їх продати на вторинному ринку з додатковою знижкою 0%, 2%, 3% 4% для кожного автомобіля відповідно. Необхідно застосувати знижку **secondary_market_discount**, як властивість для продажу кожного автомобіля на вторинному ринку, за умови, що його ціна більше 10000. 

```python
class Auto:
    
    def __init__(self, brand, current_price, discont):
        
        self.brand = brand
        self.current_price = current_price
        self.discont = discont
        self.custom_price = self.current_price * (1 - self.discont/100)
        

        
    @property
    def customer_price(self):
        return self.custom_price
    
    def secondary_market_discount(self, new_discount):
        
        if self.custom_price >= 10000:
            
            self.secondary_market_discount = self.custom_price * (1 - new_discount/100)
            print('Ціна автомобілля на вторинному ринку: ', self.secondary_market_discount)
        else:
            print('Даний автомобіль знижці не підлягає')
        

    
example1 = Auto("ЗАЗ", 7000, 10)
example2 = Auto("Мерседес", 100000, 5)
example3 = Auto("BMW", 110000, 5)
example4 = Auto("Ауді", 120000, 3)

print('Ціна автомобіля з урахуванням клієнтської знижки', example1.customer_price)
print('Ціна автомобіля з урахуванням клієнтської знижки', example2.customer_price)
print('Ціна автомобіля з урахуванням клієнтської знижки', example3.customer_price)
print('Ціна автомобіля з урахуванням клієнтської знижки', example4.customer_price, "\n")


example1.secondary_market_discount(0)
example2.secondary_market_discount(2)
example3.secondary_market_discount(3)
example4.secondary_market_discount(4)
```
