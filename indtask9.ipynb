```python
class Student:
    
    def __init__(self):
        self.k = 0
        
        
    def year(self):
        
        self.k = self.k + 1
        print(self.k, " рік навчання")
            
        if self.k == 5:
            print("Студент отримав диплом!")
        elif self.k > 5:
            print("Студент вже випустився з університету!")
        


student = Student()
for i in range (0,5):
    student.year()
```

```python
import math

class Triangle:
    
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c
        
        # self.h = h
        
    def perimetr(self):
        
        P = self.a + self.b + self.c
        print("Периметр трикутника дорівнює: " + str(P))
        
        
        
    def area(self):
        
        p = (self.a + self.b + self.c )/2
        s = round(math.sqrt(p * (p - self.a) * (p - self.b) * (p - self.c)), 3)
        print("Площа трикутника дорівнює: " + str(s))
        
        
        
    def type(self):
        
        if self.a == self.b == self.c:
            print("Даний трикутник рівносторонній")
        
        elif self.a == self.b != self.c:
            print("Даний трикутник рівнобедрений")
                  
        elif self.c ** 2 == self.b ** 2 + self.a ** 2:
            print("Даний трикутник прямокутний")
            
        else: print("Це звичайний трикутник")
            
        
    def angles(self):
        
        self.acos_a =round(math.degrees(math.acos((self.b ** 2 + self.c ** 2 - self.a ** 2)/(2*self.b * self.c))), 3)
        
        print("Величина кута а дорівнює", self.acos_a)
        
        
        self.acos_b = round(math.degrees(math.acos((self.a ** 2 + self.c ** 2 - self.b ** 2)/(2*self.a*self.c))), 3)
        print("Величина кута b дорівнює", self.acos_b)
                  
        self.acos_c = round(180 - (self.acos_a + self.acos_b), 3)
        print("Величина кута c дорівнює", self.acos_c , "\n")
        
        
triangle = Triangle(5, 8, 10)
triangle.area()
triangle.perimetr()
triangle.type()
triangle.angles()

triangle1 = Triangle(3, 4, 5)
triangle1.area()
triangle1.perimetr()
triangle1.type()
triangle1.angles()

triangle2 = Triangle(3, 3, 3)
triangle2.area()
triangle2.perimetr()
triangle2.type()
triangle2.angles()
```

```python
import math

class Quadrangle():
    
    def __init__(self, d1, d2, corner):
        
        self.d1 = d1;
        self.d2 = d2;
        self.corner = corner;
    
    def area(self):
        
        s = round((self.d1 * self.d2 * math.sin(math.radians(self.corner)))/2, 3)
        print("Площа дорівнює: " + str(s))
    
        
    def get_sides(self):
        return self.a, self.b, self.c, self.d

    
    def set_sides(self, a, b, c, d):
        self.a = a
        self.b = b
        self.c = c
        self.d = d
            
   
    
    def perimetr(self):
        p = round(self.a + self.b + self.c + self.d, 3)
        print("Периметр дорівнює: " + str(p))


class Rectangle(Quadrangle):
    
    def __init__(self, d1, corner):
        d2 = d1
        Quadrangle.__init__(self, d1, d2, corner)



    def side_a(self):

        self.a = round(self.d1 * math.sqrt((1- math.cos(math.radians(self.corner)))/2), 3)
        return self.a
        
        
    def side_b(self):
        
        self.b = round(math.sqrt( self.d1**2 - self.a**2), 3)
        return self.b
        
    
    def perimetr(self):
        
        self.c = self.a = self.side_a()
        self.d = self.b = self.side_b()
        Quadrangle.perimetr(self)
        

print("Прямокутник №1")
rectangle1 = Rectangle(10, 45)
rectangle1.area()
print("Сторона а дорівнює :", rectangle1.side_a())

print("\nПрямокутник №2")
rectangle2 = Rectangle(15, 90)
rectangle2.perimetr()
print("Сторона b дорівнює :", rectangle2.side_b())
```
