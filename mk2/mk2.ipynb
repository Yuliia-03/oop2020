```python

from pizza import *
from order import *
from bank_account import MyAccount

menu = [Pizza_Hawaii_4(), Pizza_Сapricose_5(), Pizza_Neapolitan_7(), Pizza_with_tuna_12(), Pizza_Hunting_14()]
type_list = list()

account = MyAccount('Богак', 'Юлія')
account.put_account(1000)
order = Order()

while True:
    
    print('\n№0 - Завершити роботу')
    print('№1 - Зробити замовлення')
    print('№2 - Підтвердити замовлення та оплатити рахунок')
    print('№3 - Анулювати замовлення')
    
    
    i = int(input('Виберіть операцію: '))
    
    if i == 1:
        
        print('Меню: ')
        
        for i in menu:
            print('\nНомер: {}, \nНазва: {}, \nВага: {}, \nЦіна: {}, \nІнгредієнти: {}\n'
                  .format(i.number, i.name, i.weight, i.price, i.ingredients))
        while True:
            
            my_choice = int(input('Ваш вибір піци: '))
        
            if my_choice not in range(1, len(menu)+1):
                print('Піци під даним номером не існує!')
            else:
                break
        
        print('Типи піц')
        print('№1 - Стандартна')
        print('№2 - Маленька')
        print('№3 - Велика')
        
        while True:
            pizza_type = int(input('Виберіть тип піци: '))
            
            if pizza_type not in range(1, 4):
                print('Піци даного типу не існує!')
            else:
                break
        
        order.add_order(my_order(menu[my_choice-1], pizza_type))
        if menu[my_choice-1].name not in type_list:
            type_list.append(menu[my_choice-1].name)
        order.check()
    
    elif i == 2:
        
        if len(type_list) < 3:
            print('Ви повинні замовити не менше 3 видів піц!')
        elif account.amound_of_money < order.total_price:
            print('На Вашому рахунку: ', account.amound_of_money)
            print('У Вас недостатньо коштів на рахунку! Ваш рахунок анульований!')
            order = Order()
            type_list = list()
        else:
            print('На Вашому рахунку: ', account.amound_of_money)
            print('Сума замовлення: ', order.total_price)
            account.get_accaunt(order.total_price)
            print('\nДякуємо за замовлення! Смачного!\nСтатус Вашого банківського акаунту:\n')
            print(account.my_open_account())
            order = Order()
            type_list = list()
        
    elif i == 3:
        print('Ваше замовлення було анульоване')
        order = Order()
        type_list = list()
    elif i == 0:
        print('Роботу завершено!')
        break
    else:
        print('Даної операції не існує!')
```
