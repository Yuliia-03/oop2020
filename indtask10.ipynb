```python
class groups:
    
    def __init__(self, group_name, subjects, students):
        
        self.group_name = group_name
        self.subjects = subjects
        self.students = students
        
    def change_group_name(self, group_name):
        
        self.group_name = group_name
        
    def change_subjects_number(self, subjects):
        
        self.subjects = subjects
        
    def change_students_number(self, students):
        
        self.students = students
        
        
    def print_all(self):
        
        print('Назва освітньої програми:', self.group_name)
        print('Кількість предметів у семестрі:', self.subjects)
        print('Кількість студентів на потоці:', self.students, '\n')
        
        
group1 = groups('КН', 6, 120)
group2 = groups('ПП', 7, 43)
group3 = groups('АнД', 5, 90)

group1.print_all()
group2.print_all()
group3.print_all()
print('\n')

group1.change_group_name('ІПЗ')
group2.change_subjects_number(6)
group3.change_students_number(100)

group1.print_all()
group2.print_all()
group3.print_all()
```

```python
class Books:
    
    def change_title(self, title):
        self.title = title
        
    def change_year(self, year):
        self.year = year
        
    def change_font(self, font):
        self.font = font
        
    def print_all(self):
        
        print('Назва книги: ', self.title)
        print('Рік видачі: ', self.year)
        print('Стиль шрифту: ', self.font, '\n')
        
book1 = Books()
book1.title = 'Грозовий перевал'
book1.year = 2000
book1.font = 'Times New Roman'

book1.print_all()

book2 = Books()
book2.title = 'Портрет Доріана Грея'
book2.year = 1995
book2.font = 'Georgia'

book2.print_all()

book3 = Books()
book3.title = 'Анна Кареніна'
book3.year = 2010
book3.font = 'Academy'

book3.print_all()

book1.change_title('Буремний перевал')
book2.change_year(1990)
book3.change_font('Garamond')

book1.print_all()
book2.print_all()
book3.print_all()
```

```python
class Room:
    
    def __init__(self, x, y, z, wall_color, floor_color):
        self.x = x
        self.y = y
        self.z = z
        
        self.wall_color = wall_color
        self.floor_color = floor_color
        
    def print_wall(self):
        
        self.wall = Wall(self.x, self.y, self.z, self.wall_color)
        self.wall.win_door(1, 2, 1, 2)
        self.wall.wallpainting()
        self.wall.print_all()
        print('\n')
        
    def print_floor(self):
       
        self.floor = Floor(self.x, self.y, self.floor_color).print_all()
        print('\n')
        
    def print_ceiling(self):
        
        
        self.ceiling = Сeiling(self.x, self.y)
        self.ceiling.chandelier()
        self.ceiling.painted_ceiling()
        self.ceiling.print_all()
        print('\n') 

# клас стін
class Wall:
    def __init__(self, x, y, z, color):
        self.wall = 2 * z * (x + y)
        self.color = color
        
    def win_door(self, d, e, f, g, m = 3, n = 2):
        self.window = Win_Door(d,e)
        self.door = Win_Door(f,g)
        self.numb_w = m
        self.numb_d = n
        
    def wallpainting(self):
        self.wallpainting = self.wall - self.window.square * self.numb_w - self.door.square * self.numb_d
            
    def print_all(self):
        print ("Площа стін кімнати дорівнює: " ,str(self.wall)," кв.м")
        print ("Пофарбована площа дорівнює: ", str(self.wallpainting), " кв.м")
        print('Колір фарби для стін: ', self.color)
            
# визначення площі вікон та дверей
class Win_Door:
    def __init__(self, x, y):
        self.square = x * y

# клас підлоги
class Floor:
    def __init__(self, x, y, color):
        self.floor = x * y
        self.color = color
    
    def print_all(self):
        print('Площа пофарбованої підлоги: ', self.floor, ' кв.м')
        print('Колір фарби для підлоги: ', self.color)
    
# клас стелі
class Сeiling:
    
    def __init__(self, x, y, color = 'White'):
        self.ceiling = x * y
        self.color = color
        
    def chandelier(self, a = 1, b = 1, n = 1):
        self.chandelier = Chandelier(a, b)
        self.ceiling_number = n
    
    def painted_ceiling(self):
        
        self.painted_ceiling = self.ceiling - self.chandelier.square * self.ceiling_number
        
    def print_all(self):
        
        print ("Площа стелі кімнати дорівнює:  ", str(self.ceiling)," кв.м")
        print ("Пофарбована площа дорівнює: ", str(self.painted_ceiling), " кв.м")
        print('Колір фарби для стелі: ', self.color)
        
# клас люстра
class Chandelier:
    
    def __init__(self, x, y):
        self.square = x * y
    

Class_Room = Room(10, 20, 15, 'Grey', 'Brown')
Class_Room.print_wall()
Class_Room.print_floor()
Class_Room.print_ceiling()
```
